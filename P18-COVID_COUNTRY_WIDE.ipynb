```python
# removing existing outliers from the dataset
from scipy.stats import zscore
z_score=abs(zscore(df))
print(df.shape)
df_new=df.loc[(z_score<3).all(axis=1)]
print(df_new.shape)
```

```python
df_new.skew()
```

```python
df_new['Total Tests']=boxcox(df_new['Total Tests'],0)
```

```python
df_new['Total Cases']=np.sqrt(df_new['Total Cases'])
```

```python
df_new['Total Deaths']=np.sqrt(df_new['Total Deaths'])
```

```python
df_new['Total Recovered']=boxcox(df_new['Total Recovered'],0)
```

```python
df_new['Serious or Critical']=boxcox(df_new['Serious or Critical'],0)
```

```python
df_new['Active Cases']=np.sqrt(df_new['Active Cases'])
```

```python
cols=list(df_new.columns.values)
cols.pop(cols.index('Total Deaths'))
df_new=df_new[cols+['Total Deaths']]
```

```python
#seperating my input and output variables
x=df_new.iloc[:,0:-1]
x.head()
```

```python
y=df_new.iloc[:,-1]
y.head()
```

```python
y=y.values.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_new)
x=pd.DataFrame(x,columns=df_new.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=66,test_size=0.22)
```

```python
import sklearn
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
lm = LinearRegression()
def maxr2_score(lm,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.24)
        lm=LinearRegression()
        lm.fit(x_train,y_train)
        y_pred=lm.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2_score corresponding to random state; ',r_state,' is: ',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state

    print('max r2 score corresponding to ',final_r_state,' is ',max_r_score)
maxr2_score(lm, x,y)
```

```python
#final model
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=80,test_size=0.20)
lm=linear_model.LinearRegression()
lm.fit(x_train,y_train)
y_pred=lm.predict(x_test)
```

```python
#using cross validation to check over/under fitting
from sklearn.model_selection import cross_val_score
a_score=cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring="r2")
a_score
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('r2_score: ',r2_score(y_test,y_pred))
```
