```python
def clean_dataset(df_new):
    assert isinstance(df_new, pd.DataFrame)
    df_new.dropna(inplace=True)
    indices_to_keep = ~df_new.isin([np.nan, np.inf, -np.inf]).any(1)
    return df_new[indices_to_keep].astype(np.float64)
```

```python
df_new.fillna(-99999, inplace=True)
```

```python
df_new.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
cols=list(df_new.columns.values)
cols.pop(cols.index('Claim Amount'))
df_new=df_new[['Claim Amount']+cols]
```

```python
#seperating my input and output variables
x=df_new.iloc[:,1:21]
x.head()
```

```python
y=pd.DataFrame(df_new['Claim Amount'])
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_new)
x=pd.DataFrame(x,columns=df_new.columns)
```

```python
import sklearn
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
lm = LinearRegression()
def maxr2_score(lm,x,y):
    max_r_score=0
    final_r_state = 0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.24)
        lm=LinearRegression()
        lm.fit(x_train,y_train)
        y_pred=lm.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2_score corresponding to random state; ',r_state,' is: ',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state

    print('max r2 score corresponding to ',final_r_state,' is ',max_r_score)
# calling the function 
maxr2_score(lm,x,y)
```

```python
#final model
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=80,test_size=0.20)
lm=linear_model.LinearRegression()
lm.fit(x_train,y_train)
y_pred=lm.predict(x_test)
```

```python
#using cross validation to check over/under fitting
from sklearn.model_selection import cross_val_score
a_score=cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring="r2")
a_score
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('r2_score: ',r2_score(y_test,y_pred))
```
